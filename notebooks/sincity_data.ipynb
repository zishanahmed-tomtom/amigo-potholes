```
from azure.storage.blob import BlockBlobService
import logging

def init_blob_service(acc_name, acc_key):
    logging.info("Azure blob service initialization started")
    global block_blob_service
    block_blob_service = BlockBlobService(account_name=acc_name, account_key=acc_key)
    logging.info("Azure blob service initialization completed")

def get_blobs_metadata():
    logging.info("Azure blobs metadata retrieval started")
    blobs = block_blob_service.list_blobs("models", delimiter='')
    logging.info("Azure blobs metadata retrieval completed")
    return blobs

def get_blobs_metadata_by_prefix(prefix_experiment, prefix_boundingbox):
    filtered_blobs = []
    logging.info("Azure blobs metadata retrieval started")
    blobs = block_blob_service.list_blobs("models", prefix='{0}/box={1}/'.format(prefix_experiment, prefix_boundingbox), delimiter='_')
    for blob in blobs:
        if 'part' in blob.name:
            filtered_blobs.append(blob)
    logging.info("Azure blobs metadata retrieval completed")
    return filtered_blobs

def get_blob(blob_metadata):
    # logging.info("Azure blob {0} retrieval started".format(return_blob_name(blob_metadata.name)))
    blob = block_blob_service.get_blob_to_text('models', blob_metadata.name)
    # logging.info("Azure blob {0} retrieval completed".format(return_blob_name(blob_metadata.name)))
    return blob

def get_blob_bytes(blob_metadata):
    blob = block_blob_service.get_blob_to_bytes('models', blob_metadata.name)
    return blob

def get_blob_content(blob):
    return blob.content

def return_blob_name(name):
    #experiment_1/box=118/id=8266636f-a50f-4fb7-b229-3bca02f131dc-1532714309000/part-00160-tid-7755245703421080624-92c12c94-07db-4444-b4e7-57bd2f275613-22917-1.c000.csv
    return name.split('id=')[1].split('/')[0]

def write_blob_to_csv(blob_metadata, index, coordinate):
    blob = get_blob(blob_metadata)
    logging.info("Azure blob {0} write to csv started".format(return_blob_name(blob_metadata.name)))
    text_file = open("./input/input{0}/{1}.csv".format(index, return_blob_name(blob_metadata.name)), "w")
    blob_content = get_blob_content(blob)
    text_file.write(blob_content)
    text_file.close()
    logging.info("Azure blob {0} write to csv completed".format(return_blob_name(blob_metadata.name)))
    logging.info("Azure blob {0} verification started".format(return_blob_name(blob_metadata.name)))
    result_verification = verify_blob(blob_content, coordinate)
    if result_verification:
        logging.info(bcolors.OKGREEN + "Azure blob {0} verification completed".format(return_blob_name(blob_metadata.name)) + bcolors.ENDC)
    else:
        logging.info(bcolors.FAIL + "Azure blob {0} verification failed".format(return_blob_name(blob_metadata.name)) + bcolors.ENDC)
  
```
